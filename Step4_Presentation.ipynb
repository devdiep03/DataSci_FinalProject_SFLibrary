# *Cleaning the Data sf_library


##*Filtering:

###- Standardizing Data Formats:

The "Age Range" column contained inconsistent formats, including ranges and single values. We used regular expressions to extract numerical values. If two values were present, we averaged them. If only one value was present, we used that. Missing values were filled with -1.

### - Removing Duplicates:

We checked for duplicate rows and removed them to ensure data integrity.

### - Handling Missing Values:

For "Circulation Active Year," missing values were filled with -1.

## * Aggregating:

### - Quantitative Data:

### - Age Range:

We calculated average ages for each "Patron Type" as follows:

Adult: Mean age of valid entries.

Juvenile: Mean age of valid entries.

Teen: Mean age of valid entries.

Senior: Mean age of valid entries.

We filled missing ages based on the mean value of the corresponding "Patron Type."

## *Categorical Data:

### - Home Library Definition:

We applied one-hot encoding to convert categorical values into binary indicators.

The "Home Library Code" was dropped after encoding to avoid redundancy.

### - Circulation Active Month:

We mapped the month names to numerical representations based on the order of occurrences.

### - Unnecessary Columns:

We dropped irrelevant columns, including "Patron Type Definition" and unnamed columns, to reduce noise.

#Outcome:

The cleaned dataset was saved as sf_library_cleaned_data.csv and split into training and testing sets for further analysis.
